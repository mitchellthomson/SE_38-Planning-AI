# MinMax
    MinMax in a very simple term is the idea of having the player look at future moves to attempt a way to MINIMIZE moves or tendencys that would lose a game and MAXIMIZE any opportunity possible to win the game.

    MinMax is able to plan its moves by using a "search tree" method (represented in the diagram below) where in the root node is the current state at the beginning of the current players turn and every branch is a possible move they could make, followed by the branches of those being any singular move the opponent could take, repeated until you reach the leaf of the tree.Each node along this tree is given a score as it is evaluated, in my implementation each potential move is checked to see if it would either win, lose or tie the game, once all options are examined the move with the highest given score will be played. MinMax works well for tictactoe or other games that are considered "Zero sum" as when playing these games any time a player maximizes their chance to win it is technically also minimizing the opponents chance to win, so in a sense maxing and minimizing are reaching the same goal. Each turn with MinMax one player is the maximizer (in the diagram it is the x player) and one player is the minimizer(the o player).
    
![Alt text](minimaxDiagram.png)

    MinMax is a fine method almost sort of bruteforce as it is rather time consuming to go through every possible move and every possible path down the tree all the way to the leaf nodes. This works fine for standard tictactoe as it can ge through the options (9!) quick enough, but imagine a game like chess where instead of played on a 3x3 grid it is 8x8 and you now also have to deal with complexity of having potentially at any given time 16 pieces per side of the board. This is why using MinMax it is important to reduce the depth at which you search your gamestates as well there are improvements that can be added ontop like alphabeta pruning or "Shannon type B" which is a variation where you get much faster processing of potential moves but at greater risk of missing higher scoring moves.

# AlphaBeta Pruning

    AlphaBeta Pruning is an improvement apon MinMax that takes a similar idea from Shannon type B however instead of culling at such a early stage it is able to traverse a little deeper first before pruning branches. MinMax losses alot of time analyzing bad moves so instead values are assigned to each node during evaluation (an alpha value and a beta value). These values are there to be an indicator at that node the best possible move the player could make as well as the worst possible move. The maximizing player has an understanding that everytime they make a move the next move by their opponent will be an attempt to minimize what was just done, so any branches that could allow the minimizing player to make better moves are pruned away and any pruned branches are not analyzed by the search. It is more useful this way because it will arrive to the same outcome that the standard MinMax would but can do it much faster, if properly implemented AlphaBeta will be the square root of time from Minmax.

    Its biggest flaw is not having a good understanding of exactly where to begin searching. So it can often stutter and take still a long time to compute as it randomly searches from the starting moves down, meaning it has a high potential to start by looking at a "bad move" first when if it was able to look at a high score node right away the bad move node would have been pruned and not have spent so much time.

# Monte Carlo
    Monte Carlo is a method that allows better traversal of larger trees. It is a different method than the Minmax methods above, where Minmax uses the tree and builds up each new node off a new possible move, Monte Carlo instead simulates many many games randomly to decide where to move your piece next. To break it down further, the current gameboard is cloned, and on this cloned board as many simulated games (in my case I did 200 as tictactoe is not that complicated) as you would like to run are done on this cloned gameboard. Once a simulated game is completed the starting move of that sim is recorded, at which point the cloned board is discarded, where a new board is cloned and this repeats as many times as it is designed to run (again 200 for my case). After every simulation is completed all the starting moves are analyzed to see which one lead to the most wins of the sims this is the move the AI will then make.

    This leads for some good advantages, it is really good solution for a complex game, chess and tictactoe are not the most complex as chess each piece only has set movement patterns and tictactoes scope is not that large. Something like GO where you have a large board and game pieces that can essentially go anywhere on the board would have massive move trees if you used MinMax, where with the Monte Carlo solution you dont really store these large move trees as while it is running it is just keeping your actual gameboard on its own and then playing a seperate game on the clone board which will be massive memory saving. MCTS is around a O(1) complexity for space where if you had the 4x4 grid tictactoe it would be O(7) using a standard tree (minmax)

    It does have some issues, even when running thousands of simulations because it is all random moves the sims make there is the posibility that it makes the wrong move because the "perfect" move wasnt played enough in the simulations, so while it is a good memory saver and time saver it may not always be the most accurate

    There is an extenstion of Monte Carlo called Monte Carlo Tree Search (MCTS). MCTS helps deal with this inaccuracy, after the simulations are run the possible moves are given values of probibility based on how likely they would be a winning move. Starting from the node the tree is then searched in highest probibility of winning, this can help remove some of the inaccuracy from just going off random simulations, and it retains rather fast speeds for the tree search because most of the time the first child of the node with the highest probibilty of winning will lead down the path to the best possible move.
